```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
netflix = pd.read_csv('viewings.csv')
```

```python
netflix
```

```python
show_titles = list(netflix['Show Title'])
```

```python
top_views = pd.Series(show_titles).value_counts().nlargest(10)
```

```python
n = len(top_views)
s = np.arange(n)

colors = plt.get_cmap('viridis')

plt.figure(figsize=(10,5))
plt.bar(top_views.index, top_views.values,
       color=colors(s/n))

plt.ylabel("Freq", fontsize=12)
plt.xlabel("Show Titles", fontsize=12)
plt.xticks(rotation=30, ha='right', fontsize=11)

plt.title('My Top 10 Shows Based on Viewing Activity', fontsize=16)


plt.show()
```
